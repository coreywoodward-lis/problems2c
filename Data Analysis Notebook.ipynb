```
import pandas as pd
import matplotlib.pyplot as plt
```

```
# loading local search results
local_search_smooth = pd.read_csv('firm_local_search_smooth_v2.csv')
local_search_medium = pd.read_csv('firm_local_search_medium_v2.csv')
local_search_complex = pd.read_csv('firm_local_search_complex_v2.csv')
```

```
# loading learning results
learning_search_smooth = pd.read_csv('firm_learning_search_smooth_v2.csv')
learning_search_medium = pd.read_csv('firm_learning_search_medium_v2.csv')
learning_search_complex = pd.read_csv('firm_learning_search_complex_v2.csv')
```

```
# loading war-room results
war_room_search_smooth = pd.read_csv('firm_war-room_search_smooth_v2.csv')
war_room_search_medium = pd.read_csv('firm_war-room_search_medium_v2.csv')
war_room_search_complex = pd.read_csv('firm_war-room_search_complex_v2.csv')
```

```
landscapes = ['K=0', 'K=5', 'K=11']
average_performance_improvements = [
    local_search_smooth.IMPROVEMENT.mean(),
    local_search_medium.IMPROVEMENT.mean(),
    local_search_complex.IMPROVEMENT.mean()
]

average_performance = [
    local_search_smooth['PERFORMANCE'].mean(),
    local_search_medium['PERFORMANCE'].mean(),
    local_search_complex['PERFORMANCE'].mean()
]


# Create Plot

fig, ax1 = plt.subplots() 
  
ax1.set_xlabel('Landscape Complexity') 
ax1.set_ylabel('Average Performance Improvement', color = 'orangered') 
ax1.plot(landscapes, average_performance_improvements, color = 'orangered') 
ax1.tick_params(axis ='y', labelcolor = 'orangered') 
  
# Adding Twin Axes

ax2 = ax1.twinx() 
  
ax2.set_ylabel('Average Performance', color = 'mediumorchid') 
ax2.plot(landscapes, average_performance, color = 'mediumorchid') 
ax2.tick_params(axis ='y', labelcolor = 'mediumorchid') 
 
# Show plot

plt.title("LOCAL SEARCH")


```
